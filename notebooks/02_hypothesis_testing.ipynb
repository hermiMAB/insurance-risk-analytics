Hypothesis 1: Risk differences across provincesNull Hypothesis ($H_0$): There is no significant difference in average Claim Severity (TotalClaims) between provinces.Alternative Hypothesis ($H_1$): There is a statistically significant difference in average Claim Severity between provinces.Test Used: Pairwise T-Test (with Bonferroni correction for multiple comparisons).Group Definitions:Group A (Control): Gauteng (Baseline province with the highest volume of policy data).Group B (Test): KwaZulu-Natal (Comparison category).Statistical Result: $p = 0.000656$Decision: Reject $H_0$ (since $p < 0.05$).Business Interpretation: We reject the null hypothesis for provinces ($p < 0.001$). Our analysis demonstrates that policyholders in KwaZulu-Natal exhibit a significantly higher average claim severity—specifically 33.1% higher—than those in our control group, Gauteng. Recommended Action: We must implement regional risk-rating factors in our premium pricing engine to explicitly account for the higher financial liability observed in the KwaZulu-Natal region.

Our analysis identifies a highly significant association between geographical zones and claim frequency, proving that risk is not uniformly distributed across the country. Recommended Action: We must move away from flat national pricing models. The underwriting team should immediately transition to dynamic, zone-based base rates that accurately reflect the unique risk profile of each Cresta Zone.
